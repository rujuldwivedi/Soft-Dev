# Data Augmentation
Here we'll augment the data to handle NaN values and also show histograms, boxplots, etc. to show data relationships.

## Importing the libraries

```python
import pandas as pd 
```

## Importing the dataset

```python
player_df = pd.read_csv('player_data.csv')
```

```python
# Display the first 5 rows of the dataset
player_df.head()
```

## Handling NaN values

```python
# Display the sum of missing values in the dataset
player_df.isna().sum()
```

```python
# Display the information about the dataset, like the data types of the columns, the number of non-null values in each column, etc.
player_df.info()
```

```python
# this drop columns that have no values.
player_df.dropna(axis='columns', inplace=True, how='all')

# Display the sum of missing values in the dataset again, to confirm that the columns have been dropped
player_df.isna().sum()
```

```python
# Display the information about the dataset again, to confirm that the columns have been dropped
player_df.info()
```

```python
# this drop rows that have no values.
player_df.dropna(inplace=True, how='all')

# Display the sum of missing values in the dataset again, to confirm that the rows have been dropped
player_df.isna().sum()
```

```python
# Display the information about the dataset again, to confirm that the rows have been dropped
player_df
```

```python
# this drop all rows that don't have at least 12 non-NaN values.
player_df.dropna(inplace=True, thresh=12)

# Display the sum of missing values in the dataset again, to confirm that the rows have been dropped
player_df.isna().sum()
```

```python
# Display the information about the dataset again
player_df.head(10)
```

```python
# Renumber the DataFrame index to account for the dropped rows.
player_df.reset_index(drop=True, inplace=True)

# Display the information about the dataset this time to confirm that the rows have been dropped and the index has been renumbered
player_df.info()
```

```python
# Display the summary statistics of the dataset
player_df.describe()
```

## Plotting box plots

```python
# Plotting the box plots for all the numeric columns in the dataset, which is a good way to visualize the distribution of the data and identify outliers

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter out non-numeric columns from cols
cols = player_df.columns
numeric_cols = [col for col in cols if pd.api.types.is_numeric_dtype(player_df[col])]

# Calculate the number of subplots needed dynamically based on numeric columns
num_cols = 5
num_rows = len(numeric_cols) // num_cols + (1 if len(numeric_cols) % num_cols else 0)

# Create a matrix of subplots for numeric columns only
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 11), squeeze=False)  # Ensure axes is always 2D
fig.tight_layout(pad=2.0)

# Loop over the numeric columns of the DataFrame and create a box plot for each one
for i, col in enumerate(numeric_cols):
    sns.boxplot(y=player_df[col], ax=axes[i // num_cols, i % num_cols])

# If there are any remaining empty subplots, hide them
for j in range(i + 1, num_rows * num_cols):
    axes[j // num_cols, j % num_cols].set_visible(False)

# Display the plots
plt.show()
```

```python
# Identify the index number of the row that has the lowest value in 'gls'.
goals_outlier = player_df['Gls'].idxmin()

# Display the row with the lowest value in 'gls'
goals_outlier
```

```python
# Identify the index number of the row that has the lowest value in 'MP'.
MatchPlayed_outlier = player_df['MP'].idxmin()

# Display the row with the lowest value in 'MP'
MatchPlayed_outlier
```

```python
# Drop the row that has the outlying values for 'goals'.
player_df.drop(player_df.index[goals_outlier], inplace=True)

# Check the end of the DataFrame to ensure that the correct row was dropped.
player_df.tail(15)
```

```python
# Renumber the DataFrame index to reflect the dropped rows.
player_df.reset_index(drop=True, inplace=True)
```

```python
# Display the summary statistics of the dataset again to confirm that the row has been dropped
player_df.isna().sum()
```

## Plotting histograms

```python
# Here we'll create a histogram for each numeric column in the dataset to visualize the distribution of the data.

import math

# Calculate the number of rows and columns dynamically.
num_cols = 5
num_rows = math.ceil(len(cols) / num_cols)

# Create the subplots with adjusted height ratios for more space between rows.
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 11), gridspec_kw={'height_ratios': [3]*num_rows})

# Loop over the columns in the DataFrame and create a histogram for each one.
for i in range(len(cols)):
    ax = axes[i // num_cols, i % num_cols]
    ax.hist(player_df[cols[i]], bins=30)
    ax.set_title(cols[i])

# Adjust the spacing between subplots.
plt.tight_layout(h_pad=3.0)

# Display the plots.
plt.show()
```

```python
# Create a histogram for the 'MP' column.
plt.hist(player_df['MP'], bins=30);
```

```python
# Create a histogram for the 'GP' column, this time as a probability density.
plt.hist(player_df['MP'], density=True, bins=30);
```

```python
# Plot the KDE for 'MP' over the probability-density histogram.
plt.hist(player_df['MP'], density=True, bins=15)
plt.title('MP histogram')
sns.kdeplot(player_df['MP']);
```

## Plotting histograms, but as a probability density

```python
# Create a histogram for the 'Gls' column., this time as a probability density.

import math

# Filter out non-numerical columns
numerical_cols = [col for col in player_df.columns[1:] if player_df[col].dtype in ['int64', 'float64']]

# Calculate the number of subplots needed
num_plots = len(numerical_cols)
num_rows = math.ceil(num_plots / 5)  # Adjust the number of rows based on the number of columns.

# Create a matrix of subplots and add padding around them for readability.
fig, axes = plt.subplots(num_rows, 5, figsize=(18, 4 * num_rows))
fig.tight_layout(pad=4.0)  # Adjust the padding here for better readability.

# Loop over the numerical columns of the DataFrame and create a KDE plot for each one.
for i, col in enumerate(numerical_cols):
    sns.kdeplot(data=player_df[col], ax=axes[i // 5, i % 5])

# Display the plots.
plt.show()
```

```python
# Plot the KDE for 'Gls' over the probability-density histogram.
plt.hist(player_df['Gls'], density=True, bins=15)
plt.title('Gls histogram')
sns.kdeplot(player_df['Gls']);
```
